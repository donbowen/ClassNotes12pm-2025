## Step 1: Get a dataset of firms in our sample (S&P 500 )

```python
import pandas as pd
import os

os.makedirs('inputs',exist_ok=True) 
sp500_fpath = 'inputs/SP500.csv'

# don't download if you already have it!

if not os.path.exists(sp500_fpath):
    url = 'https://raw.githubusercontent.com/datasets/s-and-p-500-companies/refs/heads/main/data/constituents.csv'
    firms = pd.read_csv(url)
    firms.to_csv(sp500_fpath, index=False)

    # alt method: via wiki (same table!)
    # url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    # pd.read_html(url)[0]

firms = pd.read_csv(sp500_fpath)
```
